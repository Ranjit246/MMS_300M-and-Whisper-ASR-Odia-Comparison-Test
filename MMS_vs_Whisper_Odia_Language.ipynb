# Audio Resampling to 16khz

```python
w,sr = torchaudio.load("/content/drive/MyDrive/test_audio/odia_i_2.wav")
w,sr
```

```python
import os
import librosa
import soundfile as sf

# Specify the path to the input folder
input_folder = '/content/drive/MyDrive/test_audio'

# Specify the path to the output folder
output_folder = '/content/drive/MyDrive/test_16_audio'

# Resampling rate
target_sr = 16000

# Iterate over the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.wav'):  # Assuming the files are in WAV format
        # Load the audio file
        input_file = os.path.join(input_folder, filename)
        audio, sr = librosa.load(input_file)

        # Resample the audio to the target sample rate
        resampled_audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)

        # Create the output file path
        output_file = os.path.join(output_folder, filename)

        # Save the resampled audio to the output folder
        sf.write(output_file, resampled_audio, target_sr)
```
